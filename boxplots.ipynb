```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv") #.drop("Unnamed: 0", axis=1)

diamonds.head()
```

```python
sns.boxplot(x="price", color="red", data=diamonds)
```

```python
#visualize the distrubition of a variable by plotting key statistics
#its better to create in seaborn
sns.boxplot(x="clarity", y="price", data=diamonds)
```

```python
sns.boxplot(x="cut", y="price", data=diamonds)
```

```python
sns.violinplot(x="cut", y="price", data=diamonds)
#they are boxplots with symmetrical kernel densities along their sides
```
