```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# can be created in seaborn with sns.barplot()
# simply specify the desired category labels and series values as "x" & "y"
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv").drop("Unnamed: 0", axis=1)

diamonds.head()
```

```python
sns.barplot(x="cut", y="price", data=diamonds);
```

```python
sns.barplot(
    x="price", # plot your numeric series as 'x' to create a horizontal bar chart
    y="cut", 
    errorbar=None, # disable error bars
    data=diamonds
);
```

```python
# grouped bar charts is better in seaborn than matplotlib
# can be created by specifying a categorical column as "hue"

# histograms can be created with sns.histplot() and a single "x" argument
# specify the number of "bins" and add the kernel density(kde=True)

#seaborn custimazation can be challenging
#histograms use seaborn
```

```python
sns.barplot(
    x="cut",
    y="carat",
    data=diamonds,
)
```

```python
# Specify the numeric series as x to create a horizontal bar chart

sns.barplot(
    x="carat",
    y="cut",
    data=diamonds,
    errorbar=None
)
```

```python
# Matplotlib formatting options generally apply to seaborn counterparts

sns.barplot(
    x="carat",
    y="cut",
    data=diamonds,
    facecolor="white",
    edgecolor="black"
)
```

```python
# Specify a categorical column in 'hue' to created grouped barplots

sns.barplot(
    x="carat",
    y="cut",
    hue="color",
    data=diamonds.sort_values(by="color"),
    palette='viridis'
)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))
sns.barplot(
    x="price",
    y="cut",
    hue="clarity",
    data=diamonds,
    palette='husl' 
)
ax.legend(bbox_to_anchor=(1.0, 1));
```

```python
sns.histplot(
    x="price", 
    bins=10, 
    kde=True, 
    stat='percent',
    data=diamonds
)
```

```python
sns.histplot(
    x="price", 
    hue="clarity",
    palette="husl",
    cbar=True,
    data=diamonds.query("clarity in ('I1', 'IF')"),
    stat='percent',
    bins=10,
    binwidth=1000,
    kde=True
)
```

```python
sns.histplot(
    x="price",
    y="clarity",
    data=diamonds,
    cbar=True #value
)
```

```python
sns.scatterplot(
    x="x",
    y="price",
    hue='carat',  
    data=diamonds,
    palette="husl",
    size="carat"
)
```

```python
sns.scatterplot(
    x="price",
    y="carat",
    hue='carat',  
    data=diamonds,
    palette="husl")
```
