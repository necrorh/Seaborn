```python
import pandas as pd
import seaborn as sns
```

```python
#pandas pivot_table method is a great way to set up the data needed for a heat map
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
```

```python
diamond_pivot = diamonds.pivot_table(
    index="cut",
    columns="color",
    values={"carat": "mean"}
)

diamond_pivot.head()
```

```python
sns.heatmap(diamond_pivot.droplevel(0,axis=1)) #remove the repeated carat
```
